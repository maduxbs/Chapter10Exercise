3.Accessor methods (getters) are defined for each attribute (e.g., get_name, get_address, etc.). These methods allow you to retrieve the values of the attributes without directly accessing the instance variables.

4.Mutator methods (setters) are defined for each attribute (e.g., set_name, set_address, etc.). These methods allow you to update the values of the attributes.

## 7 Ex.

```python
import pickle

# Employee class definition
class Employee:
    def __init__(self, emp_id, name, department, job_title):
        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.job_title = job_title

    def __str__(self):
        return f"Employee ID: {self.emp_id}, Name: {self.name}, Department: {self.department}, Job Title: {self.job_title}"

# Function to load the employee dictionary from a file
def load_employees(filename):
    try:
        with open(filename, 'rb') as file:
            return pickle.load(file)
    except FileNotFoundError:
        return {}

# Function to save the employee dictionary to a file
def save_employees(filename, employee_dict):
    with open(filename, 'wb') as file:
        pickle.dump(employee_dict, file)

# Main program
employee_data_file = "employee_data.pkl"
employee_dict = load_employees(employee_data_file)

while True:
    print("\nEmployee Management System")
    print("1. Look up an employee")
    print("2. Add a new employee")
    print("3. Change an employee's information")
    print("4. Delete an employee")
    print("5. Quit")

    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        emp_id = input("Enter Employee ID to look up: ")
        if emp_id in employee_dict:
            print(employee_dict[emp_id])
        else:
            print("Employee not found in the dictionary.")

    elif choice == '2':
        emp_id = input("Enter Employee ID: ")
        name = input("Enter Name: ")
        department = input("Enter Department: ")
        job_title = input("Enter Job Title: ")
        employee = Employee(emp_id, name, department, job_title)
        employee_dict[emp_id] = employee
        print("Employee added to the dictionary.")

    elif choice == '3':
        emp_id = input("Enter Employee ID to update: ")
        if emp_id in employee_dict:
            name = input("Enter updated Name: ")
            department = input("Enter updated Department: ")
            job_title = input("Enter updated Job Title: ")
            employee = Employee(emp_id, name, department, job_title)
            employee_dict[emp_id] = employee
            print("Employee information updated.")
        else:
            print("Employee not found in the dictionary.")

    elif choice == '4':
        emp_id = input("Enter Employee ID to delete: ")
        if emp_id in employee_dict:
            del employee_dict[emp_id]
            print("Employee deleted from the dictionary.")
        else:
            print("Employee not found in the dictionary.")

    elif choice == '5':
        save_employees(employee_data_file, employee_dict)
        print("Employee data saved. Exiting the program.")
        break

    else:
        print("Invalid choice. Please enter a valid option (1-5).")
```

## 

1.We define the Employee class to store employee information, and it has a __str__ method to provide a readable string representation of an employee.

2.We create functions to load and save the employee dictionary from/to a file using the pickle module.

3.The main program uses a while loop to present a menu to the user with options to look up, add, update, or delete employees, or to quit the program.

4.When the program starts, it loads the employee dictionary from a file (if it exists) or initializes an empty dictionary if the file is not found.

5.The program operates based on the user's input and allows them to interact with the employee dictionary.

6.When the user chooses to quit (option 5), the program saves the updated employee dictionary to the file and exits.
